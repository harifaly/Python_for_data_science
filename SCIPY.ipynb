```python
import numpy as np
import matplotlib.pyplot as plt

from scipy import optimize
def f(x):
    return np.sin(x[0])+np.cos(x[0]+x[1])*np.cos(x[0])

x=np.linspace(-3,3,100)
y=np.linspace(-3,3,100)
x,y=np.meshgrid(x,y)
plt.contour(x,y,f(np.array([x,y])),20)
x0=np.zeros(2)
plt.scatter(x0[0],x0[1],marker='+',c='r',s=100)


result=optimize.minimize(f,x0).x
print(result)

# Affichage du point trouvé
plt.scatter(result[0], result[1], c='g', s=80, label='minimum trouvé')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Avec la fonction

x = np.linspace(0,20,100)
y=x+ 4*np.sin(x)+np.random.randn(x.shape[0])
plt.plot(x,y)
```

```python
from scipy import signal
new_y=signal.detrend(y)
```

```python
plt.plot(x,y)
plt.plot(x,new_y)
```

```python
x  = np.linspace(0,30,1000)
y =  3*np.sin(x) + 2*np.sin(5*x) + np.sin(10*x)
plt.plot(x,y)
```

```python
from scipy import fftpack
```

```python
fourier = fftpack.fft(y)
frequences = fftpack.fftfreq(y.size)

plt.plot(np.abs(frequences),np.abs(fourier))
```
