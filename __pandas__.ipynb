```python
import numpy as np
```

```python
# A Pandas DataFrame is a 2 dimensional data structure, like a 2 dimensional array, or a table with rows and columns.

# --- Création ---
df = pd.DataFrame({
    'nom':    ['Alice', 'Bob', 'Clara', 'David'],
    'age':    [25, 30, np.nan, 22],
    'score':  [88, 75, 92, 60],
    'ville':  ['Paris', 'Lyon', 'Paris', 'Lyon']
})
```

```python
print(df)
```

```python
#____Exploration__
print("afficher les debuts de dataframe\n__________________________")
print(df.head())
print("___________________\nafficher les fins de dataframe\n________________________")
print(df.tail())
print("__________________\nafficher les informations de dataframe\n____________________")
print(df.info())
print("___________________\nstatistique rapide\n_____________________")
print(df.describe())
```

```python
# --- Sélection --------
print(df['score'])              #afficher une colonne
```

```python
print(df[['nom', 'score']])     #afficher  plusieurs colonnes
```

```python
print(df[df['score'] > 70])     # filtre conditionnel
```

```python
# loc sert à sélectionner des données par leurs labels (index et noms de colonnes).
# suntax:  df.loc[ligne, colonne]===> ca sert à sélectionner des données par leurs labels

print(df.loc[0, 'nom'])         # par label
```

```python
print(df.iloc[0, 1])

# Il sert à accéder aux données par position (numéro)
```

# ***GROUPBY***

sert à regrouper les données selon une colonne, puis appliquer une opération

```python
#____GroupBy______
print(df.groupby('ville')['score'].mean())   # moyenne par ville
```

```python
print(df.groupby('ville').agg({'score': ['mean', 'max', 'count']}))
```

autre operation

```python
# somme
df.groupby('ville')['score'].sum()
```

```python
# compter
df.groupby('ville')['nom'].count()
```

```python
# max
df.groupby('ville')['score'].max()
```

```python
# exemple plusieur colonnes
df.groupby('ville')[['age', 'score']].mean()
```
