```python
# A Pandas DataFrame is a 2 dimensional data structure, like a 2 dimensional array, or a table with rows and columns.

# --- Création ---
df = pd.DataFrame({
    'nom':    ['Alice', 'Bob', 'Clara', 'David'],
    'age':    [25, 30, np.nan, 22],
    'score':  [88, 75, 92, 60],
    'ville':  ['Paris', 'Lyon', 'Paris', 'Lyon']
})
```

# ***GROUPBY***

sert à regrouper les données selon une colonne, puis appliquer une opération

```python
#____GroupBy______
print(df.groupby('ville')['score'].mean())   # moyenne par ville
```

```python
print(df.groupby('ville').agg({'score': ['mean', 'max', 'count']}))
```

autre operation

```python
# somme
df.groupby('ville')['score'].sum()
```

```python
# compter
df.groupby('ville')['nom'].count()
```

```python
# max
df.groupby('ville')['score'].max()
```

```python
# exemple plusieur colonnes
df.groupby('ville')[['age', 'score']].mean()
```

# SERIES

Une Series = une colonne de données (1 dimension)

C’est comme une liste améliorée avec des index

```python
# exemple
s = pd.Series([10, 20, 30])
print(s)
print("Une Series a 2 choses:\nindex → 0, 1, 2\nvaleurs → 10, 20, 30")
```

```python
# On peut personnaliser les indexs
s = pd.Series([10, 20, 30], index=['a', 'b', 'c'])
```

```python
print(s)
```

Series = colonne unique + index

DataFrame = plusieurs Series

```python
# _______MERGE_________
# merge sert à combiner deux DataFrames selon une clé commune
```

```python
# exemple pour nos data
df2 = pd.DataFrame({'nom': ['Alice', 'Bob'], 'niveau': ['junior', 'senior']})
merged = pd.merge(df, df2, on='nom', how='left')
print(merged)
```

```python
# exemple basique
toM1 = pd.DataFrame({
    'id': [1, 2, 3],
    'nom': ['Alice', 'Bob', 'Clara']
})
```

```python
toM2= pd.DataFrame({
    'id': [1, 2, 4],
    'score': [90, 80, 70]
})
```

```python
df_merged = pd.merge(toM1, toM2, on='id')
print(df_merged)
```
