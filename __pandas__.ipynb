```python
# faut importer le package pandas
import pandas as pd
```

```python
import numpy as np
```

```python
# A Pandas DataFrame is a 2 dimensional data structure, like a 2 dimensional array, or a table with rows and columns.

# --- Création ---
df = pd.DataFrame({
    'nom':    ['Alice', 'Bob', 'Clara', 'David'],
    'age':    [25, 30, np.nan, 22],
    'score':  [88, 75, 92, 60],
    'ville':  ['Paris', 'Lyon', 'Paris', 'Lyon']
})
```

```python
#____Exploration__
print("afficher les debuts de dataframe\n__________________________")
print(df.head())
print("___________________\nafficher les fins de dataframe\n________________________")
print(df.tail())
print("__________________\nafficher les informations de dataframe\n____________________")
print(df.info())
print("___________________\nstatistique rapide\n_____________________")
print(df.describe())
```

```python
# --- Sélection --------
print(df['score'])              #afficher une colonne
```

```python
print(df[['nom', 'score']])     #afficher  plusieurs colonnes
```

```python
print(df[df['score'] > 70])     # filtre conditionnel
```

```python
# loc sert à sélectionner des données par leurs labels (index et noms de colonnes).
print(df.loc[0, 'nom'])         # par label
```

```python
print(df.iloc[0, 1])
```
