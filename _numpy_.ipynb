```python
# RESUME RAPIDE
# --- Création de tableaux ---
a = np.array([1, 2, 3])                  # vecteur 1D
A = np.array([[1, 2], [3, 4]])           # matrice 2D
Z = np.zeros((3, 3))                     # matrice de zéros
I = np.eye(3)                            # matrice identité
R = np.random.randn(4, 4)               # matrice aléatoire (loi normale)

```

```python
print("1-\n",a)
print("2-\n",A)
print("3-\n",Z)
print("4-\n",I)
print("5-\n",R)
```

```python
# --- Opérations de base ---
b = np.array([4, 5, 6])
print(a + b)        # addition élément par élément
print(a * b)        # multiplication élément par élément (PAS matricielle)
print(a @ b)        # produit scalaire (dot product)
print(A @ A)        # multiplication MATRICIELLE
```

```python
# --- Slicing ---
M = np.arange(16).reshape(4, 4)
print(M[0, :])      # première ligne
print(M[:, 1])      # deuxième colonne
print(M[1:3, 1:3])  # sous-matrice
```

```python
# --- Broadcasting (très important en ML) ---
v = np.array([1, 2, 3, 4])
print(M + v)        # ajoute v à chaque ligne de M automatiquement
```

```python
# --- Fonctions utiles ---
print(A.T)          # transposée
print(A.shape)      # dimensions
print(np.linalg.inv(A))   # inverse
print(np.linalg.det(A))   # déterminant
print(np.mean(M, axis=0)) # moyenne par colonne
print(np.std(M))          # écart-type
```
