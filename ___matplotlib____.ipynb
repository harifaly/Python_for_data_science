```python
import matplotlib.pyplot as plt
import numpy as np

```

```python
#___Graphique de base___
x = np.linspace(-3, 3, 100)
y = x**2

plt.figure(figsize=(8, 5))
plt.plot(x, y, color='steelblue', linewidth=2, label='f(x) = x²')
plt.title('Parabole')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# ___ Scatter plot ___
X = np.random.randn(100, 2)
labels = (X[:, 0] + X[:, 1] > 0).astype(int)
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='coolwarm', alpha=0.7)
plt.title('Deux classes séparables')
plt.show()
```

```python
# _______Subplots (très utile pour ML) _______
fig, axes = plt.subplots(1, 3, figsize=(14, 4))
axes[0].hist(np.random.randn(1000), bins=30, color='steelblue')
axes[0].set_title('Distribution normale')
axes[1].plot(np.cumsum(np.random.randn(200)), color='coral')
axes[1].set_title('Random walk')
axes[2].imshow(np.random.rand(10, 10), cmap='viridis')
axes[2].set_title('Heatmap')
plt.tight_layout()
plt.show()
```
